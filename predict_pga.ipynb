```python
import os
import obspy
import numpy as np
import pandas as pd
```

```python
pga_list = {"year": [],
            "julday": [],
            "pga": []}
for folder in sorted(os.listdir("Final_java_earthquake_data")):
    inv = obspy.read_inventory(f"Final_java_earthquake_data/{folder}/stations/{os.listdir(f"Final_java_earthquake_data/{folder}/stations")[0]}")
    event_pga = []
    for file in os.listdir(f"Final_java_earthquake_data/{folder}/waveforms"):
        tr = obspy.read(f"Final_java_earthquake_data/{folder}/waveforms/{file}")[0]
        tr_copy = tr.copy()
        tr_copy.remove_response(inventory=inv, output="ACC")
        pga = max(abs(tr_copy.data))
        date = tr.stats.starttime
        event_pga.append(pga)
        print(f"date: {date}, pga: {pga}")
    pga_list["year"].append(date.year)
    pga_list["julday"].append(date.julday)
    pga_list["pga"].append(np.average(event_pga))
    print()
print(pga_list)
```

```python
pga_data = pd.DataFrame(pga_list)
print(pga_data)
```
